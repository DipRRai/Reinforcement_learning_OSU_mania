```python
import gymnasium as gym
import numpy as np
from gymnasium import spaces
import random
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
class CustomEnv(gym.Env):
    """Custom Environment that follows gym interface."""

    metadata = {"render_modes": ["human"], "render_fps": 30}

    def __init__(self):
        super().__init__()
        self.action_space = spaces.Discrete(3)
        self.observation_space = spaces.Box(low=np.array([0]), high=np.array([100]), dtype=np.int32)
        self.state = 38 + random.randint(-3,3)
        self.shower_length = 60

    def step(self, action):
        self.state += action - 1
        self.shower_length -= 1 
        
        if 37 <= self.state <= 39: 
            reward = 1 
        else: 
            reward = -1

        info = {}
        
        terminated = self.shower_length <= 0
        truncated = False 
        
        return np.array([self.state], dtype=np.int32), reward, terminated, truncated, info

    def reset(self, seed=None, options=None):
        self.state = 38 + random.randint(-3, 3)
        self.shower_length = 60 
        info = {}
        return np.array([self.state]), info

    def render(self):
        pass

    def close(self):
        pass
```

```python
# Instantiate the env
env = CustomEnv()
# Define and Train the agent
model = PPO("MlpPolicy", env, verbose=1)
```

```python
model.learn(total_timesteps=100000)
```

```python
from stable_baselines3.common.env_checker import check_env

env = CustomEnv()
# It will check your custom environment and output additional warnings if needed
check_env(env)
```

```python
evaluate_policy(model, env, n_eval_episodes=10, render=False)
```
